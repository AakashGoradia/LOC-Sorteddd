```python
import tensorflow as tf
import numpy as np
```

```python
import base64
from PIL import Image

# Open image file
with open('D:\pyth\python\LOC-Sorteddd\img_1.jpg', 'rb') as image_file_1:
    img_1 = image_file_1.read()
    
with open('D:\pyth\python\LOC-Sorteddd\img_2.jpg', 'rb') as image_file_2:
    img_2 = image_file_2.read()
```

```python
base64_string_1 = base64.b64encode(img_1).decode('utf-8')
base64_string_2 = base64.b64encode(img_2).decode('utf-8')
```

```python
print(base64_string_1)
print(base64_string_2)
```

```python
import io
```

```python
# decode the first base64 image
img_data_1 = base64.b64decode(base64_string_1)
img_1 = Image.open(io.BytesIO(img_data_1))

# decode the second base64 image
img_data_2 = base64.b64decode(base64_string_2)
img_2 = Image.open(io.BytesIO(img_data_2))
```

```python
# load pre-trained CNN model
model = tf.keras.applications.ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# preprocess images
img_1 = img_1.resize((224, 224))
img_1 = np.array(img_1)
img_1 = np.expand_dims(img_1, axis=0)
img_1 = tf.keras.applications.resnet_v2.preprocess_input(img_1)

img_2 = img_2.resize((224, 224))
img_2 = np.array(img_2)
img_2 = np.expand_dims(img_2, axis=0)
img_2 = tf.keras.applications.resnet_v2.preprocess_input(img_2)
```

```python
features_1 = model.predict(img_1)
features_2 = model.predict(img_2)

# remove the first dimension of the features arrays
features_1 = np.squeeze(features_1)
features_2 = np.squeeze(features_2)

# flatten the feature vectors
flat_features_1 = np.reshape(features_1, (-1,))
flat_features_2 = np.reshape(features_2, (-1,))

# compute the dot product of the feature vectors
similarity_score = np.dot(flat_features_1, flat_features_2) / (np.linalg.norm(flat_features_1) * np.linalg.norm(flat_features_2))
print(similarity_score)
```

```python
similarity_score = ((similarity_score / 2) + 0.5) * 100  # convert to percentage score
print("Similarity score: {:.2f}%".format(similarity_score))
```
